```python
# split Data
x = df.drop('Attrition',axis=1)
y = df['Attrition']
```

```python
# Standardscalar

from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
scalar.fit_transform(x)
```

# Model Building

```python
# Splitting Data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Algorithm

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')  # Use balanced class weights to handle class imbalance
model

 
```

```python
model.fit(x_train,y_train)
```

```python
# Model Prediction

ypred_train = model.predict(x_train)
ypred_test = model.predict(x_test)
```

# Model Evaluation

```python
# Taking Calssification Report

from sklearn.metrics import classification_report

class_report = classification_report(ypred_train,y_train)
print(class_report)
```

```python
class_report = classification_report(ypred_test,y_test)
print(class_report)
```
